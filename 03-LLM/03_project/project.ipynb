## Step1

### Loading the given Data with `pandas`

```python
import os
import pandas as pd
import numpy as np
```

```python
#character_descriptions_folder_path = "./data/character_descriptions"

character_descriptions_text_raw_path = "./data/character_descriptions/character_descriptions.csv"
character_descriptions_text_clean_path = "./data/internal/character_descriptions_clean.csv"
character_descriptions_embeddings_path = "./data/internal/character_descriptions_embeddings.csv"
character_descriptions_distances_sorted_path = "./data/internal/character_descriptions_distances_sorted_path.csv"

question = "Who was up coming musical perfomer from USA, full of energy and enthusiasm?"
# Answer Johnny, row number 22
```

```python
df = pd.read_csv(character_descriptions_text_raw_path)
print(df.shape)
df
```

```python
# Clean up
df['Name'].replace('', np.nan, inplace=True)
df['Description'].replace('', np.nan, inplace=True)
df['Medium'].replace('', np.nan, inplace=True)
df['Setting'].replace('', np.nan, inplace=True)


df.dropna(subset=['Name'], inplace=True)
df.dropna(subset=['Description'], inplace=True)
df.dropna(subset=['Medium'], inplace=True)
df.dropna(subset=['Setting'], inplace=True)
```

```python
# Combine the fields for embeddings
df["text"] = (
    df["Name"].fillna("") + ", " +
    df["Description"].fillna("") + ", " +
    df["Medium"].fillna("") + ", " +
    df["Setting"].fillna("")
)

df_text = df.drop(['Name', 'Description', 'Medium', 'Setting'], axis=1)
df_text.to_csv(character_descriptions_text_clean_path)
```

### Creating an Embeddings Index with `openai.Embedding`

```python
df = pd.read_csv(character_descriptions_text_clean_path, index_col=0)
```

```python
import openai
openai.api_key = "YOUR API KEY"
```

```python
EMBEDDING_MODEL_NAME = "text-embedding-ada-002"

response = openai.Embedding.create(
    input=df["text"].tolist(),
    model=EMBEDDING_MODEL_NAME
)
```

```python
# embeddings = [data["embedding"] for data in response["data"]]

embeddings = [item.embedding for item in response.data]
df["embeddings"] = embeddings
df.to_csv(character_descriptions_embeddings_path)
df
```

## Step2

### Finding Relevant Data with Cosine Similarity

```python
import numpy as np
import pandas as pd

df_distance = pd.read_csv(character_descriptions_embeddings_path, index_col=0)
df_distance["embeddings"] = df_distance["embeddings"].apply(eval).apply(np.array)
```

```python
import openai
from openai.embeddings_utils import get_embedding




question_embeedings = get_embedding(question, engine=EMBEDDING_MODEL_NAME)
question_embeedings
```

```python
from openai.embeddings_utils import distances_from_embeddings
distances = distances_from_embeddings(question_embeedings, df["embeddings"].tolist(), distance_metric="cosine")
distances
```

```python
df_distance["distances"] = distances
df_distance
```

```python
df_distance_sorted = df_distance.sort_values(by="distances", ascending=True).reset_index(drop=True)
df_distance_sorted.to_csv(character_descriptions_distances_sorted_path)
df_distance_sorted
```

### Tokenizing with `tiktoken`

```python
import tiktoken
```

```python
tokenizer = tiktoken.get_encoding("cl100k_base")
```

### Basic prompt

```python
basic_prompt_template = """
Answer the question based on the context below, and if the question
can't be answered based on the context, say "I don't know"

Context: 

{}

---

Question: {}
Answer:"""
```

```python
print(basic_prompt_template.format("context", question))
```

### Custom prompt

```python
custom_prompt_template = """
You are an expert research assistant skilled in extracting detailed, evidence-based answers from text. 
Your goal is to provide an accurate, well-structured answer grounded ONLY in the given context.

Follow these rules:
1. Read the context carefully and identify the most relevant information.
2. If the question is ambiguous, state your assumptions before answering.
3. If the context lacks enough information, respond with: "I don't know based on the given context."
4. When possible, summarize key reasoning steps in 23 sentences.
5. Present the final answer clearly under an **Answer:** section.

Context:
{}

---

Question: {}
Answer:
"""
```

```python
print(custom_prompt_template.format("context", question))
```

```python
max_token_count = 1000
current_token_count = len(tokenizer.encode(basic_prompt_template)) + len(tokenizer.encode(question))
current_token_count
```

```python
context = []
df_distance_sorted = pd.read_csv(character_descriptions_distances_sorted_path, index_col=0)
```

```python
for text in df_distance_sorted["text"].values:
    text_token_count = len(tokenizer.encode(text))
    current_token_count += text_token_count
    
    if current_token_count <= max_token_count:
        context.append(text)
    else:
        break
```
